```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the Excel file
def load_excel_file(file_path):
    try:
        data = pd.read_excel(file_path)
        return data
    except Exception as e:
        print("Error loading Excel file:", str(e))

# Generate plots
def generate_plots(data):
    # Histogram: Price Distribution
    plt.figure(figsize=(8, 6))
    sns.histplot(data['Price'], bins=10)
    plt.title('Price Distribution')
    plt.xlabel('Price')
    plt.ylabel('Frequency')
    plt.show()

    # Pie Chart: Genre-wise Book Count
    plt.figure(figsize=(8, 6))
    genre_counts = data['Genre'].value_counts()
    plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%')
    plt.title('Genre-wise Book Count')
    plt.show()

    # Linear Plot: User Rating vs Price
    plt.figure(figsize=(8, 6))
    sns.regplot(x='Price', y='User Rating', data=data)
    plt.title('User Rating vs Price')
    plt.xlabel('Price')
    plt.ylabel('User Rating')
    plt.show()

    # Seaborn Plots: Scatter Plot, Bar Plot, Box Plot
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x='Price', y='User Rating', data=data)
    plt.title('Scatter Plot: User Rating vs Price')
    plt.xlabel('Price')
    plt.ylabel('User Rating')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.barplot(x='Genre', y='Price', data=data)
    plt.title('Bar Plot: Price by Genre')
    plt.xlabel('Genre')
    plt.ylabel('Price')
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Genre', y='Price', data=data)
    plt.title('Box Plot: Price by Genre')
    plt.xlabel('Genre')
    plt.ylabel('Price')
    plt.show()



    # Scatter Plot: User Rating vs Price
    plt.figure(figsize=(8, 6))
    plt.scatter(data['Price'], data['User Rating'])
    plt.title('Scatter Plot: User Rating vs Price')
    plt.xlabel('Price')
    plt.ylabel('User Rating')
    plt.show()

# Main function
def main():
    file_path = 'amazon_books.xlsx'  # Replace with your file path
    data = load_excel_file(file_path)

    if data is not None:
        generate_plots(data)

if __name__ == "__main__":
    main()

```
